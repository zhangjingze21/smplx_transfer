```python
def compare_dicts(dict1, dict2):
    # 检查键值对数量
    if len(dict1) != len(dict2):
        return False

    # 检查所有键是否相同
    if set(dict1.keys()) != set(dict2.keys()):
        return False

    # 检查所有键对应的值是否相同
    for key in dict1:
        if dict1[key] != dict2[key]:
            return False

    return True

# 示例字典
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'a': 1, 'b': 2, 'c': 3}

# 比较两个字典
if compare_dicts(dict1, dict2):
    print("两个字典相同")
else:
    print("两个字典不同")
```
